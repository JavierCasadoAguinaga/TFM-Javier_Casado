```python
def time_frame (value):
    
    if ((value >= datetime.time(7,0)) & (value < datetime.time(14,0))):
        return 'Morning'
    if ((value >= datetime.time(14,0)) & (value < datetime.time(16,0))):
        return 'Lunch'
    if ((value >= datetime.time(16,0)) & (value < datetime.time(20,30))):
        return 'After'
    if ((value >= datetime.time(20,30)) & (value <= datetime.time(23,59))):
        return 'PT'
    if ((value >= datetime.time(0,0)) & (value < datetime.time(0,30))):
        return 'PT'
    if ((value >= datetime.time(0,30)) & (value < datetime.time(1,30))):
        return 'Late_PT'
    if ((value >= datetime.time(1,30)) & (value < datetime.time(7,0))):
        return 'Night'
    
#df1['TM_Bis'] = df1.HR.swifter.apply(lambda valor: op_time_frame(valor))
```
