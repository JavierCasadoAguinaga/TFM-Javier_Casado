```python
#ELIMINAMOS LAS CABECERAS DE LOS PROGRAMAS
df.dropna(inplace = True)
df = df[df.Descripción.str.startswith('<<')]
```

```python
#ELIMINAMOS LOS SIMBOLOS <> DEL DATA FRAME
df1 = pd.DataFrame()
for i in df.columns:
    df1 = pd.concat([df1,df.loc[:,str(i)].str.translate({ord(x): None for x in '<>'})], axis = 1)
```

```python
#DIVIDIMOS LA COLUMNA EN DOS PARA PODER OPERAR SOLO CON LA HORA 
#(NO SE PUEDE PASAR A DATETIME YA QUE NO ESTÁ EN EL RANGO ADECUADO)
df2 = pd.DataFrame()
df2 = pd.concat([df2,df1], axis = 1)
df2[['Hora','Minuto']] = df2.Descripción.str.split(':', expand = True)
```

```python
#SE CAMBIA EL TIPO DE DATO EN LAS COLUMNAS HORA MINUTO PARA PODER TRANSFORMAR A RANGO 00:00-23:00(STR-INT)
df2 = df2.astype({'Hora':'int','Minuto':'int'})
```

```python
#SE RESTA DOS HORAS A 
df2.Hora = df2['Hora'].map(lambda value : value - 2)
df2.Hora = df2.Hora.replace(24,0)
```

```python
#REEMPLAZAMOS LAS COMAS DE LOS VALORES NUMÉRICOS POR PUNTOS PARA PODER CONVERTIRLOS A FLOAT
df2[['Ind. 4+ (inv)','Amas de Casa (c/inv.)', 'Ind. 4 - 12 (c/inv.)','Ind. 16+ (c/inv.)','Ind. 4+ (inv).1',
           'Amas de Casa (c/inv.).1','Ind. 4 - 12 (c/inv.).1', 'Ind. 16+ (c/inv.).1']] = df2[['Ind. 4+ (inv)','Amas de Casa (c/inv.)', 'Ind. 4 - 12 (c/inv.)','Ind. 16+ (c/inv.)','Ind. 4+ (inv).1',
           'Amas de Casa (c/inv.).1','Ind. 4 - 12 (c/inv.).1', 'Ind. 16+ (c/inv.).1']].stack().str.replace(',','.').unstack()
```

```python
#CONVERTIMOS A FLOAT EL RESTO DE VALORES NÚMERICOS QUE SE EXTRAEN COMO STR(HACER AL CARGAR LOS DATOS: DTYPE)
df2 = df2.astype({'Ind. 4+ (inv)':'float', 'Amas de Casa (c/inv.)':'float', 'Ind. 4 - 12 (c/inv.)':'float',
       'Ind. 16+ (c/inv.)':'float', 'Ind. 4+ (inv).1':'float', 'Amas de Casa (c/inv.).1':'float',
       'Ind. 4 - 12 (c/inv.).1':'float', 'Ind. 16+ (c/inv.).1':'float'})
```

```python
#VOLVEMOS A CONVERTIR HORA Y MINUTO A STR PARA PODER UNIRLOS CON ':' COMO SEPARADOR
df2 = df2.astype({'Hora':'str', 'Minuto':'str'})
```

```python
#UNIMOS HORA Y MINUTO PARA DESPUÉS UNIRLO A LA FECHA Y CONVERTIRLO EN TIMESTAMP
df2 ['Tiempo'] = df2[['Hora', 'Minuto']].apply(lambda x: ':'.join(x), axis=1)
df2 ['Fecha_comp'] = df2[['Fecha', 'Tiempo']].apply(lambda x: ' '.join(x), axis=1)
df2.Fecha_comp = pd.to_datetime(df2.Fecha_comp)
```

```python
#ELIMINAMOS LAS COLUMNAS GENERADAS Y USADAS PARA LAS TRANSFORMACIONES
df2 = df2.drop(['Tiempo','Hora', 'Minuto','Descripción','Fecha'], axis = 1)
```
